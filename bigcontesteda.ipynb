```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
from preprocess import *
```

```python
data = pd.read_csv("2023빅콘테스트_어드밴스드리그_예술의전당.csv")
```

```python
for column in data.columns:
    unique_values = data[column].unique()
    unique_value_count = data[column].nunique()
    print(f"For column {column}, unique values are {unique_values}")
    print(f"Number of unique values: {unique_value_count}")
```

```python
# age 컬럼
total_age = len(data)
nan_age = data['age'].isna().sum()
non_nan_age = total_age - nan_age
ratio_age = non_nan_age / total_age

print(f"Age: NaN = {nan_age}, Not NaN = {non_nan_age}, 비율 = {ratio_age:.4f}")

# member_yn 컬럼
total_member_yn = len(data)
nan_member_yn = data['member_yn'].isna().sum()
non_nan_member_yn = total_member_yn - nan_member_yn
ratio_member_yn = non_nan_member_yn / total_member_yn

print(f"Member_yn: NaN = {nan_member_yn}, Not NaN = {non_nan_member_yn}, 비율 = {ratio_member_yn:.4f}")
```

```python
# member_yn 컬럼에서 'Y'와 'N'의 개수
count_member_yn = data['member_yn'].value_counts()

# 전체 member_yn 컬럼의 행 수 (NaN 제외)
total_member_yn = count_member_yn.sum()

# 각 값의 비율 계산
ratio_y = count_member_yn['Y'] / total_member_yn
ratio_n = count_member_yn['N'] / total_member_yn

print(f"'Y'의 비율: {ratio_y:.2f}")
print(f"'N'의 비율: {ratio_n:.2f}")
```

```python
# age가 NaN인 행을 추출하고 복사본을 만듭니다.
age_nan_df = data[data['age'].isna()].copy()

# extract_age 함수 적용
age_nan_df['age'] = age_nan_df['discount_type'].apply(extract_age)

# 원본 DataFrame에서 age가 NaN인 행을 제거
data = data.dropna(subset=['age'])

# age가 수정된 DataFrame과 원본 DataFrame을 병합
data = pd.concat([data, age_nan_df]).sort_index()

# 결과 출력
print(data)
```

```python
# age 컬럼
total_age = len(data)
nan_age = data['age'].isna().sum()
non_nan_age = total_age - nan_age
ratio_age = non_nan_age / total_age

print(f"Age: NaN = {nan_age}, Not NaN = {non_nan_age}, 비율 = {ratio_age:.4f}")

# member_yn 컬럼
total_member_yn = len(data)
nan_member_yn = data['member_yn'].isna().sum()
non_nan_member_yn = total_member_yn - nan_member_yn
ratio_member_yn = non_nan_member_yn / total_member_yn

print(f"Member_yn: NaN = {nan_member_yn}, Not NaN = {non_nan_member_yn}, 비율 = {ratio_member_yn:.4f}")
```

```python
import numpy as np

# discount_type 별로 age의 평균을 계산
age_means = data.groupby('discount_type')['age'].mean()

# 10의 자리에서 반올림
age_means_rounded = age_means.round(-1)

# age가 NaN인 행을 추출
age_nan_rows = data['age'].isna()

# age의 결측치를 채우기 위해 discount_type 별로 평균 age를 매핑
data.loc[age_nan_rows, 'age'] = data.loc[age_nan_rows, 'discount_type'].map(age_means_rounded)
```

```python
# age 컬럼
total_age = len(data)
nan_age = data['age'].isna().sum()
non_nan_age = total_age - nan_age
ratio_age = non_nan_age / total_age

print(f"Age: NaN = {nan_age}, Not NaN = {non_nan_age}, 비율 = {ratio_age:.4f}")

# member_yn 컬럼
total_member_yn = len(data)
nan_member_yn = data['member_yn'].isna().sum()
non_nan_member_yn = total_member_yn - nan_member_yn
ratio_member_yn = non_nan_member_yn / total_member_yn

print(f"Member_yn: NaN = {nan_member_yn}, Not NaN = {non_nan_member_yn}, 비율 = {ratio_member_yn:.4f}")
```

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder

# 범주형 변수를 숫자로 변환
le = LabelEncoder()
data['gender'] = le.fit_transform(data['gender'].astype(str))  # 'gender' 라벨인코딩
data['member_yn'] = le.fit_transform(data['member_yn'].astype(str))  # 'member_yn' 라벨인코딩

# 원래의 'gender'와 'member_yn'을 복사해 둡니다.
original_gender = data['gender'].copy()
original_member_yn = data['member_yn'].copy()
```

```python
# KNNImputer 객체 생성 (k=3과 n_jobs=-1로 설정)
imputer = KNNImputer(n_neighbors=3, n_jobs=-1)

# 'age', 'gender', 'member_yn' 컬럼에 적용하고 결과를 해당 컬럼에 다시 저장
imputed_data = imputer.fit_transform(data[['age', 'gender', 'member_yn']])

# 'age' 컬럼만 수정합니다.
data['age'] = imputed_data[:, 0]

# 'gender'와 'member_yn'을 원래 값으로 복원합니다.
data['gender'] = original_gender
data['member_yn'] = original_member_yn
```
